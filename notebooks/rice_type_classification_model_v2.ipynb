```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader, random_split
from sklearn.model_selection import train_test_split
from torch.utils.data import Subset
from collections import Counter
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

### Image transformation

```python
train_transform = transforms.Compose([ # data preprocessing
    #transforms.Resize((320, 320)),
    transforms.Resize(384),
    transforms.CenterCrop(320),
    transforms.RandomHorizontalFlip(), # augmentations -> prevent overfitting
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

val_test_transform = transforms.Compose([
    #transforms.Resize((320, 320)),
    transforms.Resize(384),
    transforms.CenterCrop(320),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

### load dataset

```python
data_dir = "/Users/vijani/Documents/hoogschool/block B/projects/RiceScan/dataset"

full_dataset = datasets.ImageFolder(
    root=data_dir,
    transform=val_test_transform  # temporary, will override for train
)

class_names = full_dataset.classes
num_classes = len(class_names)

print("Classes:", class_names)

targets = full_dataset.targets
indices = list(range(len(full_dataset)))
```

### Stratified split: Train vs Temp (70% / 30%)

```python
train_indices, temp_indices = train_test_split(
    indices,
    test_size=0.30,
    stratify=targets,
    random_state=42
)
```

### Stratified split: Validation vs Test (15% / 15%)

```python
temp_targets = [targets[i] for i in temp_indices]

val_indices, test_indices = train_test_split(
    temp_indices,
    test_size=0.50,
    stratify=temp_targets,
    random_state=42
)
```

### Create PyTorch subsets

```python
train_dataset = Subset(full_dataset, train_indices)
val_dataset   = Subset(full_dataset, val_indices)
test_dataset  = Subset(full_dataset, test_indices)
```

### assign transforms

```python
train_dataset.dataset.transform = train_transform
val_dataset.dataset.transform = val_test_transform
test_dataset.dataset.transform = val_test_transform

```

### Create data loaders

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

### verify split balance: stratification

```python
# def count_labels(subset):
#     labels = [full_dataset.targets[i] for i in subset.indices]
#     return Counter(labels)

# print("Train:", count_labels(train_dataset))
# print("Val:", count_labels(val_dataset))
# print("Test:", count_labels(test_dataset))

def count_labels(indices):
    labels = [targets[i] for i in indices]
    return Counter(labels)

print("Train:", count_labels(train_indices))
print("Val:", count_labels(val_indices))
print("Test:", count_labels(test_indices))
```

### Model: Transfer Learning (MobileNetV2)

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = models.mobilenet_v2(pretrained=True)

# Freeze backbone
for param in model.features.parameters():
    param.requires_grad = False # feature extraction - This "freezes" the pre-trained layers, meaning the model keeps the sophisticated edge-detection and shape-recognition skills it learned on ImageNet.

# Replace classifier
model.classifier[1] = nn.Linear( # final layer is replaced with a new nn.Linear layer that maps the extracted features to the 4 specific rice classes.
    model.last_channel,
    num_classes
)

model = model.to(device)
```

### Loss function and optimizer

```python
criterion = nn.CrossEntropyLoss() # for multiclass classification

optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=0.001 # learning rate
)
```

### Training loop

```python
def train_model(model, train_loader, val_loader, epochs=15):
    train_losses, val_losses = [], []
    train_accuracies, val_accuracies = [], []

    for epoch in range(epochs):
        # ---- Training ----
        model.train()
        running_loss = 0.0
        correct, total = 0, 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        train_loss = running_loss / len(train_loader)
        train_acc = correct / total

        # ---- Validation ----
        model.eval()
        val_loss = 0.0
        correct, total = 0, 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item()
                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)

        val_loss /= len(val_loader)
        val_acc = correct / total

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accuracies.append(train_acc)
        val_accuracies.append(val_acc)

        print(f"Epoch [{epoch+1}/{epochs}] "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

    return train_losses, val_losses, train_accuracies, val_accuracies
```

### Train model

```python
train_losses, val_losses, train_accs, val_accs = train_model(
    model, train_loader, val_loader, epochs=20
)
```

### Plot Loss + Accuracy in ONE FIGURE

```python
epochs_range = range(1, len(train_losses) + 1)

plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(epochs_range, train_losses, label="Train Loss")
plt.plot(epochs_range, val_losses, label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training & Validation Loss")
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs_range, train_accs, label="Train Accuracy")
plt.plot(epochs_range, val_accs, label="Val Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training & Validation Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

### Evaluation on test set (accuracy)

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct, total = 0, 0
    y_true, y_pred = [], []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)

            correct += (preds == labels).sum().item()
            total += labels.size(0)

            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    test_acc = correct / total
    print(f"Test Accuracy: {test_acc:.4f}")

    return y_true, y_pred
```

```python
y_true, y_pred = evaluate_model(model, test_loader)
```

### Confusion matrix

```python


cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_names
)

fig, ax = plt.subplots(figsize=(8, 6))  # increase figure size if needed
disp.plot(cmap="Blues", values_format="d", ax=ax)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')  # rotate 45 degrees, align right
plt.yticks(rotation=0)  # optional, keep y-axis labels horizontal

plt.title("Confusion Matrix")
plt.tight_layout()  # adjusts layout so labels are not cut off
plt.show()
```
